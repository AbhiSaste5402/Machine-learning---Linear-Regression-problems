```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Downloads/canada_per_capita_income.csv")
df
```

```python
df.rename(columns = {'per capita income (US$)':'income'}, inplace = True)
df
```

```python
plt.scatter(df.year, df.income)
plt.xlabel("years")
plt.ylabel("income")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']], df.income)
```

```python
reg.predict(np.array([2020]).reshape(1, 1)) # X= year you want to predict in y=mx+b
```

```python
reg.coef_   # y=mx+b  this is m
```

```python
reg.intercept_  # b
```
